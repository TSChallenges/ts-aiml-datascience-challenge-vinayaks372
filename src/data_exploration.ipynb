```python

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt         #  importing some basic libraries to work with the dataset and for visualization



def load_data(filepath):
    df = pd.read_csv(filepath);
    return df

def check_missing_values(df):
    missing = df.isnull();
    count = missing.sum();
    print("Missing Values:\n", count)

def generate_summary_statistics(df):
    summary = df.describe()
    print("Summary Statistics:\n", summary)

def visualize_distributions(df):

    column = ["age", "balance", "credit_score", "estimated_salary"]
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))
    axes = axes.flatten()

    for i, col in enumerate(column):
        sns.histplot(df[col], ax=axes[i], kde=True)
        axes[i].set_title(f'Distribution of {col}')

    plt.tight_layout()
    plt.show()

def main():
    df = load_data("/workspaces/ts-aiml-datascience-challenge-vinayaks372/data/bank_churn.csv")

    # Check for missing values
    check_missing_values(df)
    
    # Generate summary statistics
    generate_summary_statistics(df)
    
    # Visualize distributions
    visualize_distributions(df)

if __name__ == "__main__":
    main()
```
